```python
# BLOCK 1: IMPORT THÆ¯ VIá»†N & LOAD FILE TRONG VS CODE
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler

print("ğŸ”¹ BÆ¯á»šC 1: NHáº¬P ÄÆ¯á»œNG DáºªN FILE Dá»® LIá»†U (VD: data.xlsx hoáº·c C:/Users/You/data.csv):")
filename = input("Nháº­p tÃªn hoáº·c Ä‘Æ°á»ng dáº«n file:  ")

print(f"-> Äang Ä‘á»c file: {filename}")

# Äá»c file (há»— trá»£ csv/xlsx)
if filename.endswith(".csv"):
    df = pd.read_csv(filename)
elif filename.endswith(".xlsx"):
    df = pd.read_excel(filename)
else:
    raise ValueError("âŒ Chá»‰ há»— trá»£ file .csv hoáº·c .xlsx")

print("-> ÄÃ£ Ä‘á»c file thÃ nh cÃ´ng!")
print(df.head())
```

```python
# BLOCK 2: Äá»ŒC Dá»® LIá»†U TRONG VS CODE
import pandas as pd

try:
    if filename.endswith((".xlsx", ".xls")):
        df = pd.read_excel(filename)
        print("-> ÄÃ£ Ä‘á»c thÃ nh cÃ´ng Ä‘á»‹nh dáº¡ng Excel (.xlsx/.xls)")

    elif filename.endswith(".csv"):
        df = pd.read_csv(filename)
        print("-> ÄÃ£ Ä‘á»c thÃ nh cÃ´ng Ä‘á»‹nh dáº¡ng CSV")

    else:
        raise ValueError("âŒ Lá»—i: Chá»‰ há»— trá»£ file CSV hoáº·c Excel!")

except Exception as e:
    print(f"âŒ Lá»–I: KhÃ´ng Ä‘á»c Ä‘Æ°á»£c file. Chi tiáº¿t: {e}")
    raise

print("\n5 dÃ²ng Ä‘áº§u cá»§a dá»¯ liá»‡u gá»‘c:")
print(df.head())
```

```python
# BLOCK 3: CHá»ŒN Lá»ŒC Cá»˜T (FEATURE SELECTION)
print("\nğŸ”¹ BÆ¯á»šC 2: CHá»ŒN Lá»ŒC CÃC Cá»˜T QUAN TRá»ŒNG...")

# Danh sÃ¡ch cá»™t má»¥c tiÃªu (sá»­ dá»¥ng tÃªn gá»‘c 'name', 'artist')
target_columns = [
    'track_id', 'name', 'artist',
    'tempo', 'energy', 'valence', 'danceability', 'acousticness',
    'instrumentalness', 'loudness', 'speechiness',
    'tags', 'year', 'play_count'
]

# Chá»‰ láº¥y nhá»¯ng cá»™t thá»±c sá»± tá»“n táº¡i trong file
existing_cols = [col for col in target_columns if col in df.columns]
df_clean = df[existing_cols].copy()

print(f"-> ÄÃ£ giá»¯ láº¡i {len(existing_cols)} cá»™t.")
print(f"-> CÃ¡c cá»™t hiá»‡n táº¡i: {existing_cols}")
```

```python
# BLOCK 4: Xá»¬ LÃ GIÃ TRá»Š NULL
print("\nğŸ”¹ BÆ¯á»šC 3: Xá»¬ LÃ Dá»® LIá»†U KHUYáº¾T THIáº¾U...")

# XÃ¡c Ä‘á»‹nh cÃ¡c cá»™t sá»‘ cÃ³ trong dá»¯ liá»‡u
numeric_cols = ['tempo', 'energy', 'valence', 'danceability', 'acousticness',
                'instrumentalness', 'loudness', 'speechiness', 'year', 'play_count']
available_numeric = [c for c in numeric_cols if c in df_clean.columns]

# 1. Äiá»n trung bÃ¬nh (mean) cho cá»™t sá»‘
for col in available_numeric:
    if df_clean[col].isnull().any():
        mean_val = df_clean[col].mean()
        df_clean[col].fillna(mean_val, inplace=True)
        print(f"   -> ÄÃ£ fillna cá»™t '{col}'")

# 2. Chuyá»ƒn Ä‘á»•i play_count sang sá»‘ nguyÃªn (náº¿u cÃ³)
if 'play_count' in df_clean.columns:
    df_clean['play_count'] = df_clean['play_count'].round(0).astype('Int64')

# 3. Xá»­ lÃ½ tÃªn bÃ i hÃ¡t/nghá»‡ sÄ© thiáº¿u
text_cols = ['name', 'artist']
for col in text_cols:
    if col in df_clean.columns:
        df_clean[col].fillna('unknown', inplace=True)

# 4. Xá»­ lÃ½ Tags (chuáº©n hÃ³a vá» chá»¯ thÆ°á»ng)
if 'tags' in df_clean.columns:
    df_clean['tags'].fillna('no_tags', inplace=True)
    df_clean['tags'] = df_clean['tags'].astype(str).str.lower()

print("-> HoÃ n táº¥t xá»­ lÃ½ Missing Data.")
```

```python
# BLOCK 5: MÃƒ HÃ“A Cá»˜T TAGS
if 'tags' in df_clean.columns:
    print("\nğŸ”¹ BÆ¯á»šC 4: MÃƒ HÃ“A Cá»˜T TAGS...")

    # TÃ¡ch chuá»—i tags thÃ nh list vÃ  Ä‘áº¿m táº§n suáº¥t
    all_tags_series = df_clean['tags'].str.split(r'[,;]\s*').explode()
    # Lá»c bá» rÃ¡c
    all_tags_series = all_tags_series[~all_tags_series.isin(['', 'no_tags', 'nan', 'unknown'])]

    # Chá»n Top 50 tags phá»• biáº¿n nháº¥t
    TOP_N = 50
    top_tags = all_tags_series.value_counts().head(TOP_N).index.tolist()
    print(f"-> Top 5 tags phá»• biáº¿n nháº¥t: {top_tags[:5]}")

    # Táº¡o cá»™t 0/1 cho má»—i tag
    for tag in top_tags:
        # Táº¡o cá»™t má»›i (VD: tag_rock)
        df_clean[f'tag_{tag}'] = df_clean['tags'].str.contains(tag, regex=False).astype(int)

    # XÃ³a cá»™t tags gá»‘c sau khi Ä‘Ã£ mÃ£ hÃ³a xong
    df_clean.drop(columns=['tags'], inplace=True)
    print(f"-> ÄÃ£ táº¡o thÃªm {TOP_N} cá»™t Ä‘áº·c trÆ°ng tá»« Tags.")
else:
    print("\nâš ï¸ Cáº£nh bÃ¡o: KhÃ´ng tÃ¬m tháº¥y cá»™t 'tags', bá» qua bÆ°á»›c nÃ y.")
```

```python
# BLOCK 6: CHUáº¨N HÃ“A (SCALING)
print("\nğŸ”¹ BÆ¯á»šC 5: CHUáº¨N HÃ“A Dá»® LIá»†U Sá»...")
scaler = MinMaxScaler()

# Láº¥y láº¡i danh sÃ¡ch cá»™t sá»‘ cáº§n scale (chá»‰ scale cÃ¡c cá»™t features gá»‘c, khÃ´ng scale cá»™t tag 0/1)
cols_to_scale = [c for c in available_numeric if c in df_clean.columns]

if cols_to_scale:
    # Thá»±c hiá»‡n scale vá» khoáº£ng [0, 1]
    df_clean[cols_to_scale] = scaler.fit_transform(df_clean[cols_to_scale])
    print(f"-> ÄÃ£ chuáº©n hÃ³a {len(cols_to_scale)} cá»™t.")
else:
    print("-> KhÃ´ng cÃ³ cá»™t sá»‘ nÃ o Ä‘á»ƒ chuáº©n hÃ³a.")

print(df_clean.head())
```

```python
# BLOCK 7: LÆ¯U FILE
output_file = 'music_data_ready_for_train_final.csv'
df_clean.to_csv(output_file, index=False)

print("\n" + "="*60)
print(f"âœ… HOÃ€N Táº¤T! File sáº¡ch Ä‘Ã£ Ä‘Æ°á»£c lÆ°u: {output_file}")
print(f"ğŸ“Š KÃ­ch thÆ°á»›c dá»¯ liá»‡u: {df_clean.shape[0]} dÃ²ng, {df_clean.shape[1]} cá»™t")
print("="*60)
```

```python
# BLOCK 8: LOAD Dá»® LIá»†U Sáº CH & CHUáº¨N Bá»Š TRAIN
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

# 1. Load file CSV vá»«a xá»­ lÃ½ xong
input_file = 'music_data_ready_for_train_final.csv'
df = pd.read_csv(input_file)

# 2. TÃ¡ch Feature (Dá»¯ liá»‡u há»c) vÃ  Metadata (ThÃ´ng tin Ä‘á»‹nh danh)
# CÃ¡c cá»™t KHÃ”NG dÃ¹ng Ä‘á»ƒ train (chá»‰ Ä‘á»ƒ hiá»ƒn thá»‹)
metadata_cols = ['track_id', 'name', 'artist']
feature_cols = [col for col in df.columns if col not in metadata_cols]

# X lÃ  ma tráº­n chá»‰ chá»©a sá»‘ (Features + Tags)
X = df[feature_cols].values

# 3. Chia táº­p Train/Test (80% train, 20% test)
# random_state=42 Ä‘á»ƒ káº¿t quáº£ giá»‘ng nhau má»—i láº§n cháº¡y
X_train, X_test = train_test_split(X, test_size=0.2, random_state=42)

print(f"âœ… Dá»¯ liá»‡u sáºµn sÃ ng!")
print(f"- Sá»‘ lÆ°á»£ng bÃ i hÃ¡t dÃ¹ng Ä‘á»ƒ train: {X_train.shape[0]}")
print(f"- Sá»‘ lÆ°á»£ng features Ä‘áº§u vÃ o (Input dim): {X_train.shape[1]}")
```

```python
# BLOCK 9: XÃ‚Y Dá»°NG MODEL AUTOENCODER
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Dropout

# KÃ­ch thÆ°á»›c Ä‘áº§u vÃ o (báº±ng sá»‘ features)
input_dim = X_train.shape[1]

# --- 1. ENCODER (NÃ©n dá»¯ liá»‡u) ---
input_layer = Input(shape=(input_dim,))
encoded = Dense(128, activation='relu')(input_layer) # NÃ©n láº§n 1
encoded = Dropout(0.2)(encoded)                      # Giáº£m overfitting
encoded = Dense(64, activation='relu')(encoded)      # NÃ©n láº§n 2

# --- 2. BOTTLENECK (Vector Ä‘áº·c trÆ°ng - QUAN TRá»ŒNG NHáº¤T) ---
# NÃ©n xuá»‘ng cÃ²n 32 chiá»u (Embedding Vector)
bottleneck = Dense(32, activation='relu', name='embedding_layer')(encoded)

# --- 3. DECODER (Giáº£i nÃ©n - Chá»‰ dÃ¹ng Ä‘á»ƒ train) ---
decoded = Dense(64, activation='relu')(bottleneck)
decoded = Dense(128, activation='relu')(decoded)
# Output layer pháº£i cÃ³ kÃ­ch thÆ°á»›c báº±ng Input layer
output_layer = Dense(input_dim, activation='sigmoid')(decoded)

# Gá»™p thÃ nh model hoÃ n chá»‰nh
autoencoder = Model(inputs=input_layer, outputs=output_layer)

# Cáº¥u hÃ¬nh thuáº­t toÃ¡n tá»‘i Æ°u (Adam) vÃ  hÃ m máº¥t mÃ¡t (MSE)
autoencoder.compile(optimizer='adam', loss='mse')

# Xem cáº¥u trÃºc máº¡ng
autoencoder.summary()
```

```python
# BLOCK 10: HUáº¤N LUYá»†N MODEL (TRAINING)
print("ğŸš€ Báº¯t Ä‘áº§u train model...")

history = autoencoder.fit(
    X_train, X_train,   # Autoencoder há»c cÃ¡ch tÃ¡i táº¡o chÃ­nh nÃ³
    epochs=20,          # Há»c 20 vÃ²ng (cÃ³ thá»ƒ tÄƒng lÃªn 50 náº¿u muá»‘n ká»¹ hÆ¡n)
    batch_size=128,     # Há»c 128 bÃ i cÃ¹ng lÃºc
    shuffle=True,
    validation_data=(X_test, X_test),
    verbose=1
)

print("\nâœ… ÄÃ£ train xong!")
```

```python
# BLOCK 11: TRÃCH XUáº¤T VECTOR & LÆ¯U FILE FINAL
from tensorflow.keras.models import Model

# 1. LÆ°u Model Autoencoder (Ä‘á»ƒ dÃ¹ng sau nÃ y náº¿u cáº§n train tiáº¿p)
autoencoder.save('music_recommender_autoencoder.h5')

# 2. TÃ¡ch láº¥y pháº§n Encoder (Bá»™ táº¡o Vector)
encoder_model = Model(inputs=autoencoder.input, outputs=autoencoder.get_layer('embedding_layer').output)
encoder_model.save('music_encoder_only.h5')
print("-> ÄÃ£ lÆ°u model Encoder: music_encoder_only.h5")

# 3. Táº¡o Vector cho TOÃ€N Bá»˜ dá»¯ liá»‡u
print("-> Äang táº¡o vector Ä‘áº·c trÆ°ng cho toÃ n bá»™ kho nháº¡c...")
song_vectors = encoder_model.predict(X)

# 4. GÃ©p Vector vÃ o láº¡i DataFrame gá»‘c (cÃ³ tÃªn bÃ i, nghá»‡ sÄ©)
# Táº¡o tÃªn cá»™t cho vector (emb_0, emb_1, ..., emb_31)
embedding_cols = [f'emb_{i}' for i in range(song_vectors.shape[1])]
df_embeddings = pd.DataFrame(song_vectors, columns=embedding_cols)

# Gá»™p dá»¯ liá»‡u: Metadata + Vector
df_final = pd.concat([df[metadata_cols], df_embeddings], axis=1)

# 5. LÆ°u file cuá»‘i cÃ¹ng (DÃ¹ng cho há»‡ thá»‘ng tÃ¬m kiáº¿m/API)
output_final_csv = 'music_vector_database.csv'
df_final.to_csv(output_final_csv, index=False)

print("\n" + "="*60)
print(f"ğŸ‰ HOÃ€N Táº¤T TOÃ€N Bá»˜ QUÃ TRÃŒNH!")
print(f"ğŸ“‚ File dá»¯ liá»‡u cuá»‘i cÃ¹ng (Metadata + Vector): {output_final_csv}")
print("Báº¡n cÃ³ thá»ƒ táº£i file nÃ y vá» Ä‘á»ƒ xÃ¢y dá»±ng API hoáº·c App.")
print("="*60)
print(df_final.head())
```

```python
# BLOCK 12: TEST Há»† THá»NG Gá»¢I Ã (INFERENCE)
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# 1. XÃ¡c Ä‘á»‹nh cÃ¡c cá»™t Vector (emb_0, emb_1, ...)
# (Giáº£ sá»­ df_final Ä‘Ã£ cÃ³ tá»« Block 11)
vector_cols = [col for col in df_final.columns if col.startswith('emb_')]
song_vectors_matrix = df_final[vector_cols].values

def recommend_song(input_name, top_n=10):
    # --- BÆ°á»›c 1: TÃ¬m bÃ i hÃ¡t trong database ---
    # TÃ¬m kiáº¿m gáº§n Ä‘Ãºng (chá»©a tá»« khÃ³a, khÃ´ng phÃ¢n biá»‡t hoa thÆ°á»ng)
    matches = df_final[df_final['name'].str.contains(input_name, case=False, na=False)]

    if matches.empty:
        print(f"âŒ KhÃ´ng tÃ¬m tháº¥y bÃ i hÃ¡t nÃ o cÃ³ tÃªn chá»©a: '{input_name}'")
        return

    # Láº¥y káº¿t quáº£ Ä‘áº§u tiÃªn tÃ¬m Ä‘Æ°á»£c lÃ m bÃ i hÃ¡t máº«u (Query Song)
    query_index = matches.index[0]
    query_song = df_final.iloc[query_index]
    query_vector = song_vectors_matrix[query_index].reshape(1, -1)

    print(f"\nğŸ” Äang tÃ¬m bÃ i hÃ¡t tÆ°Æ¡ng tá»± vá»›i:\n   ğŸµ BÃ i hÃ¡t: {query_song['name']}\n   ğŸ¤ Nghá»‡ sÄ©: {query_song['artist']}")

    # --- BÆ°á»›c 2: TÃ­nh Ä‘á»™ tÆ°Æ¡ng Ä‘á»“ng (Cosine Similarity) ---
    # So sÃ¡nh vector cá»§a bÃ i máº«u vá»›i Táº¤T Cáº¢ cÃ¡c bÃ i khÃ¡c
    similarity_scores = cosine_similarity(query_vector, song_vectors_matrix).flatten()

    # --- BÆ°á»›c 3: Láº¥y Top N káº¿t quáº£ cao nháº¥t ---
    # Sáº¯p xáº¿p giáº£m dáº§n vÃ  bá» qua chÃ­nh bÃ i hÃ¡t Ä‘Ã³ (vá»‹ trÃ­ Ä‘áº§u tiÃªn)
    top_indices = similarity_scores.argsort()[::-1][1:top_n+1]

    # --- BÆ°á»›c 4: Hiá»ƒn thá»‹ káº¿t quáº£ ---
    print(f"\nâœ… Gá»¢I Ã {top_n} BÃ€I HÃT TÆ¯Æ NG Äá»’NG NHáº¤T:")
    print("-" * 80)
    print(f"{'BÃ€I HÃT':<35} | {'NGHá»† SÄ¨':<25} | {'Äá»˜ GIá»NG':<10}")
    print("-" * 80)

    for i in top_indices:
        rec_name = df_final.iloc[i]['name']
        rec_artist = df_final.iloc[i]['artist']
        score = similarity_scores[i]

        # Cáº¯t tÃªn ngáº¯n láº¡i náº¿u quÃ¡ dÃ i Ä‘á»ƒ in cho Ä‘áº¹p
        print(f"{str(rec_name)[:33]:<35} | {str(rec_artist)[:23]:<25} | {score:.4f}")
    print("-" * 80)

# --- CHáº Y THá»¬ NGAY ---
# Báº¡n cÃ³ thá»ƒ nháº­p tÃªn bÃ i hÃ¡t tiáº¿ng Anh báº¥t ká»³ (VD: "Love", "Night", "Radio", "Rock")
input_song = input("âŒ¨ï¸ Nháº­p tÃªn bÃ i hÃ¡t báº¡n muá»‘n tÃ¬m: ")
recommend_song(input_song)
```
